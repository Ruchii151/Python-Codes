```python
nums= [1,2,3,4]
output=[]
current_sum= 0

for i in nums:
    current_sum += i
    output.append(current_sum)
print(output)
```

```python
# Function for above code

def running_sum(n):
    output=[]
    current_sum= 0
    
    for i in n:
        current_sum += i
        output.append(current_sum)
    return output
```

```python
nums2= [1,1,1,1,1,1,1,1,1]
running_sum(nums2)
```

```python
nums3= [4,3,7,2,8,4,8,9,10]
running_sum(nums3)
```

```python
nos= [2,5,1,3,4,7]
n=3

op=[]

for i in range(n):
    op.append(nos[i])
    op.append(nos[i+n])
print(op)
```

```python
# function for the above code 

def shuffle_array(nos,n):
    op= []

    for i in range(n):
        op.append(nos[i])
        op.append(nos[i+n])

    return op
```

```python
nos2= [1,2,3,4,4,3,2,1]
n2=4
shuffle_array(nos2,n2)
```

```python
# Approach 1 with in built max method

candies= [2,3,5,1,3]
extra_candies= 3
maximum= max(candies)

ans= []

for i in candies:
    if i + extra_candies >= maximum:
        ans.append(True)
    else:
        ans.append(False)

print(ans)
```

```python
# Function for the above code

def extra_candies_check(candies, extr_candies):
    ans= []

    for i in candies:
        if i + extra_candies >= maximum:
            ans.append(True)
        else:
            ans.append(False)
    
    return ans
```

```python
extra_candies_check(candies, extra_candies)
```

```python
# Approach 2 without in built method 

def extra_candies_2(candies, extra_candies):
    max_candies= candies[0]
    
    for i in candies:
        if i > max_candies:
            max_candies= i
    
    
    ans= []
    
    for i in candies:
        if i+ extra_candies >= max_candies:
            ans.append(True)
        else:
            ans.append(False)
    
    return ans
```

```python
extra_candies_2(candies, extra_candies)
```
